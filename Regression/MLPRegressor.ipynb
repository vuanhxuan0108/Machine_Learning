```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

```python
data = pd.read_csv(r"Cellphone.csv")
data.head()
```

```python
dtTrain, dtTest = train_test_split(data, test_size = 0.3, shuffle = True)  
X_Train = dtTrain.iloc[:,2:]
y_Train = dtTrain.iloc[:,1]
X_Test = dtTest.iloc[:,2:]
y_Test = dtTest.iloc[:,1]
```

```python
model = MLPRegressor(hidden_layer_sizes=(170, 200,) ,max_iter=5000, activation="identity", alpha=0.001)
model.fit(X_Train, y_Train)
y_pred = model.predict(X_Test)
```

```python
print("Thuc te           Du doan                Chenh lech")
for i in range(0, len(y_Test)):
    print(y_Test.values[i], "\t", y_pred[i],"\t", y_Test.values[i]-y_pred[i])
```

```python
from sklearn.metrics import r2_score,mean_absolute_error, mean_squared_error
y_mean = np.mean(y_Test)
# TÃ­nh NSE
nse = 1 - (np.sum((y_Test - y_pred) ** 2) / np.sum((y_Test - y_mean) ** 2))

#Tinh R2
r2 = r2_score(y_Test, y_pred)

# TÃ­nh MAE
mae = mean_absolute_error(y_Test, y_pred)

# TÃ­nh RMSE
rmse = np.sqrt(mean_squared_error(y_Test, y_pred))

# In giÃ¡ trá» 
print("NSE:", nse)
print("R2: ", r2)
print("MAE:", mae)
print("RMSE:", rmse)
```
